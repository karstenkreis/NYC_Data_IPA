####Just a quick look to see whether Citibike usage and weather are related

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
import seaborn as sb
```

```python
weather = pd.read_pickle("clean_weather.pkl")
citibike = pd.read_pickle("citibike_July-Aug2013.pkl")
```

Take day 43 as a rainy day and day 50 as a dry day.

```python
bike_rain = citibike[(citibike["month"]==8) & (citibike["day_of_month"]==13)]
bike_dry = citibike[(citibike["month"]==8) & (citibike["day_of_month"]==20)]
```

```python
bike_rainy = bike_rain[["hour"]]
bike_rainy["Count"] = 0.0
```

```python
bike_rainy = bike_rainy[["hour", "Count"]].groupby(["hour"]).count()
```

```python
bike_rainy.head(2)
```

```python
bike_dry = bike_dry[["hour"]]
bike_dry["Count"] = 0.0
```

```python
bike_dry = bike_dry[["hour", "Count"]].groupby(["hour"]).count()
```

```python
bike_dry.head(2)
```

```python
bike_dry.rename(columns={'Count': 'Trips_Dry'}, inplace=True)
bike_rainy.rename(columns={'Count': 'Trips_Rainy'}, inplace=True)
```

```python
bike_both = pd.concat([bike_dry, bike_rainy], axis=1, join_axes=[bike_dry.index])
```

```python
rainy_weather = weather[weather["AbsoluteDay"]==43][["Hour","PrecipitationIn","Conditions"]]
```

```python
dry_weather = weather[weather["AbsoluteDay"]==50][["Hour","PrecipitationIn","Conditions"]]
```

```python
rainy_weather.rename(columns={'PrecipitationIn': 'PrecipitationIn_Rainy', 'Conditions': 'Conditions_Rainy'}, inplace=True)
dry_weather.rename(columns={'PrecipitationIn': 'PrecipitationIn_Dry', 'Conditions': 'Conditions_Dry'}, inplace=True)
```

```python
weather = dry_weather.merge(rainy_weather, on="Hour")
```

```python
weather.fillna(0.0)
```

```python
f = bike_both.plot()
plt.xlabel('Hour', fontsize=15)
plt.ylabel('# Trips per Hour', fontsize=15)
f = weather["PrecipitationIn_Rainy"].plot(secondary_y=True)
plt.title('Citebike Usage vs. Rain', fontsize=15)
plt.ylabel('Precipitation in Inch', fontsize=15)
plt.legend(fontsize=10, loc = "lower right")
```

In the morning there is rain and the citibike usage decreases. 
In the afternoon though, citibike usage is nearly back to normal as the rain has stopped.
Also note that the citibike usage is already lower than on a dry day before the rain has even started.
This demonstrates the correlation with weather forecasts and not only the weather itself.
